```python
import numpy as np
csv=np.genfromtxt ('./data/etkina_data.csv',delimiter=",")
csv[0:2]
```

```python
csv=csv[1:]
```

```python
csv[:2]
```

```python
t=csv[:,0]
x_ball=csv[:,1]
y_ball=csv[:,2]
x_clock=csv[:,3]
y_clock=csv[:,4]
x_rel=csv[:,5]
y_rel=csv[:,6]
```

```python
t_th = np.linspace(min(t),max(t))
fit=4.031-1.1*t_th
```

```python
from matplotlib import pyplot as plt
plt.plot(t,x_ball,"o",label="ball")
plt.plot(t,x_clock,"x",label="clock")
plt.plot(t,x_rel,"*",label="ball relative to clock")
plt.plot(t_th,fit,"-",label="-1.1m/s slope")
plt.xlabel("time (s)")
plt.ylabel("horizontal position (m)")
plt.legend()

#plt.show()
plt.savefig("etkina-x.pdf")
```
