```python
import numpy as np
csv=np.genfromtxt ('./data/bear_data.csv',delimiter=",")
csv[0:2]
```

```python
csv=csv[1:]
```

```python
csv[:2]
```

```python
t=csv[:,0]
x_bear=csv[:,1]
y_bear=csv[:,2]
x_window=csv[:,3]
y_window=csv[:,4]
x_rel=csv[:,6]
y_rel=csv[:,5]
```

```python
t_th_1 = np.linspace(min(t),16.15);
t_th_2 = np.linspace(17.25,17.7);

speed_1 = 195.9-12.29*t_th_1
speed_2 = 104.4-5.972*t_th_2
```

```python
from matplotlib import pyplot as plt
plt.plot(t,y_bear,"o",label="bear")
plt.plot(t,y_window,"x",label="window")
plt.plot(t,y_rel,"*",label="bear relative to window")
plt.plot(t_th_1,speed_1,"--",label="12.3 m/s slope")
plt.plot(t_th_2,speed_2,":",label="6.0 m/s slope")
plt.xlabel("time (s)")
plt.ylabel("vertical position (m)")
plt.legend()

#plt.show()
plt.savefig("bear_y.pdf")
```
