```python
import numpy as np
csv=np.genfromtxt ('./data/bear_data.csv',delimiter=",")
csv[0:2]
```

```python
csv=csv[1:]
```

```python
csv[:2]
```

```python
t=csv[:,0]
x_bear=csv[:,1]
y_bear=csv[:,2]
x_window=csv[:,3]
y_window=csv[:,4]
x_rel=csv[:,6]
y_rel=csv[:,5]
```

```python
t_th_1 = np.linspace(16.117,max(t))

height = -4.6930291*t_th_1*t_th_1 + 158.88985*t_th_1 - 1343.4845
```

```python
from matplotlib import pyplot as plt
plt.plot(t,y_bear,"o",label="bear")
plt.plot(t,y_window,"x",label="window")
plt.plot(t,y_rel,"*",label="bear relative to window")
plt.plot(t_th_1,height,"--",label="g = -9.4 $m/s^2$")
plt.xlabel("time (s)")
plt.ylabel("vertical position (m)")
plt.legend()

#plt.show()
plt.savefig("bear-quadratic.pdf")
```
